# Feature Selection

```python
df_features=df[['Movie_Genre', 'Movie_Language','Movie_Keywords', 'Movie_Overview','Movie_Cast','Movie_Director']].fillna('')
```

```python
df_features.head(2)
```

```python
X= df_features['Movie_Genre']+' '+df_features['Movie_Language']+' '+df_features['Movie_Keywords']+' '+df_features['Movie_Overview']+' '+df_features['Movie_Cast']+' '+df_features['Movie_Director']
```

```python
X
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
X=tfidf.fit_transform(X)
```

```python
print(X)
```

# Cosine similarity score

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(X)
```

```python
Similarity_Score
```

```python
Movie_name= input("Enter a Movie Name : ")
```

```python
Movie_Names_List = df['Movie_Title'].tolist()
```

```python
import difflib
```

# List of Similar Movies

```python
Similar_names=difflib.get_close_matches(Movie_name,Movie_Names_List)
print(Similar_names)
```

```python
Closest_match=Similar_names[0]
print(Closest_match)
```

```python
ID_of_Movie = df[df.Movie_Title==Closest_match]['Movie_ID'].values[0]
print(ID_of_Movie)
```

```python
Recommendation_Score = list(enumerate(Similarity_Score[ID_of_Movie]))
print(Recommendation_Score)
```

# Sorting Movies based on Similarity Score

```python
Sorted_Similar_Movies = sorted(Recommendation_Score, key=lambda x:x[1], reverse = True)
print(Sorted_Similar_Movies)
```

# Printing the Titles of Similar Movies

```python
print("Top 10 movies Suggested for you :")
i=1
for m in Sorted_Similar_Movies:
    index=m[0]
    title=df[df.index==index]['Movie_Title'].values[0]
    if(i<11):
        print(i, "--> ", title)
        i+=1
```

```python
Your_movie_name= input("Enter a Movie Name : ")
list_of_titles= df['Movie_Title'].tolist()
Close_match=difflib.get_close_matches(Your_movie_name,list_of_titles)[0]
Index_of_Movie = df[df.Movie_Title==Close_match]['Movie_ID'].values[0]
Recommendation_Score = list(enumerate(Similarity_Score[Index_of_Movie]))
Sorted_similar_movies = sorted(Recommendation_Score, key=lambda x:x[1], reverse = True)
print("Top 10 movies Suggested for you :")
i=1
for m in Sorted_similar_movies:
    index=m[0]
    title=df[df.index==index]['Movie_Title'].values[0]
    if(i<11):
        print(i, "--> ", title)
        i+=1
```
